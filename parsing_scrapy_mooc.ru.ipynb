```python
import scrapy
import time

class MoocSpider(scrapy.Spider):
    name = 'mooc'
    start_urls = ['https://mooc.ru/courses']

    def parse(self, response):
        for link in response.xpath('/html/body/div[7]/div/section/div/div[4]/a/@href'):
            yield response.follow(link, callback=self.parse_page)
        link = 'https://mooc.ru/courses'
        time.sleep(1)
  
    def parse_page(self, response):
        program_list = []
        for i in range(len(response.css('div#v-pills-3 div.title::text'))):
            program_item = response.css('div#v-pills-3 div.title::text')[i].get()
            program_list.append(program_item)

        key_skills = []
        for i in range(len(response.css('div#v-pills-4 div.title::text'))):
            key_skills_item = response.css('div#v-pills-4 div.title::text')[i].get().strip().replace(',',', ')
            key_skills.append(key_skills_item)
        for i in range(len(response.css('div#v-pills-5 div.title::text'))):
            key_skills_item2 = response.css('div#v-pills-5 div.title::text')[i].get().strip()
            key_skills.append(key_skills_item2)

        price = response.css('div.course-price::text').get().strip()
        if not price: price = 'free'

        credit = response.css('div#v-pills-2 p::text').get()
        if not credit: credit = ''

        duration = response.xpath('/html/body/div[7]/div/section[2]/div/div[1]/div[1]/div[2]/div[1]/div/p/text()').get()
        if not duration: duration = ''

        yield {
            'title': response.css('div.wrap h1::text').get(),
            'price': price,
            'school': response.css('div.col-12 p.d-md-block::text').get(),
            'credit': credit,
            'duration': duration,
            'program': program_list,
            'key_skills': key_skills
        }
```
